# Model Viewer Example

This example shows how to create an ESP server model viewer.

First off we import ESPPY and the Visuals

```python
import esppy
from esppy.espapi.visuals import Visuals
```

Set the server_url to point to your ESP server

```python
server_url = "http://<host>:<port>"
```

Create the ESPPY instance, the server connection, and the Visuals object.

```python
esp = esppy.ESP(server_url)
conn = esp.createServerConnection()
visuals = Visuals()
```

Load a couple projects

```python
esp.load_project("model_1.xml")
esp.load_project("model_2.xml")
```

```python
viewer = visuals.createModelViewer(conn,height="600px",cpu=True,index=True)
viewer
```
